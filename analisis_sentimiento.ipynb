```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
path = r'C:/Users/CMA1702/Documents/Maestria/proyectoAplicado/Audios_Diego_Part1_2/Audios_Diego_Part1_2/Grabaciones-1/2023-04-01a/temp/0c1dacc2-0d21-474f-9f30-0b2bc3069d1a_20230401T16_11_UTC.wav'

# Cargar el archivo de audio
sample_rate, audio_data = wavfile.read(path)

# Si el archivo de audio tiene más de un canal (estéreo), tomamos solo un canal
if len(audio_data.shape) > 1:
    audio_data = audio_data[:, 0]
```

```python
time_axis = np.linspace(0, len(audio_data) / sample_rate, num=len(audio_data))
plt.plot(time_axis, audio_data)
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.title('Señal de Audio en el Dominio del Tiempo')
```

```python
# Se Crea el espectrograma con el mapeo de colores 'viridis'
plt.specgram(audio_data, Fs=sample_rate, cmap='viridis')

# Añadir etiquetas y título
plt.xlabel('Tiempo [s]')
plt.ylabel('Frecuencia [Hz]')
plt.title('Espectrograma')

# Mostrar el espectrograma
plt.show()
```
